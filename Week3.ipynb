# **Transformasi Linier Brightness**

ùëî(ùë•, ùë¶) = ùëì(ùë•, ùë¶) + ùëè 

g(x,y) adalah nilai pixel setelah transformasi, f(x,y) adalah nilai pixel asli, dan b adalah
nilai brightness

```python
print('Mengubah tingkat kecerahan citra')
print('--------------------------------')
try:
  brightness = int(input('masukkan nilai kecerahan: '))
except ValueError:
  print('Error, not a number')


original = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/female.tiff')
brightness_image = np.zeros(original.shape, original.dtype)

#Akses per piksel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(original[y,x,c] + brightness, 0, 255)


#Cara sederhana tanpa loop
#brightness_image = cv.convertScaleAbs(original, beta=brightness)

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

**Implementasikan transformasi contrast**

```python
print('Mengubah Kontras dan Tingkat Kecerahan Citra')
print('--------------------------------------------')
try:
  brightness_level = int(input('masukkan nilai kecerahan [-255 - 255]: '))
  contrast_level = float(input('masukkan nilai kontras [1.0 - 3.0]: '))
except ValueError:
  print('Error, not a number')

image1 = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/female.tiff')
formatted_image1 = np.zeros(image1.shape, image1.dtype)

#Akses per piksel
for y in range(image1.shape[0]):
  for x in range(image1.shape[1]):
    for c in range(image1.shape[2]):
      formatted_image1[y,x,c] = np.clip(contrast_level * image1[y,x,c] + brightness_level, 0, 255)

final_frame = cv.hconcat((image1, formatted_image1))
cv2_imshow(final_frame)
```

**Implementasikan transformasi grayscale menggunakan metode averaging, lightness, dan luminance**

```python
import matplotlib.pyplot as plt

print('Implementasi Transformasi Grayscale')
print('-----------------------------------')

image3 = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/mandrill.tiff')
averaging_image3 = np.zeros(image3.shape, image3.dtype)

image3 = cv.cvtColor(image3, cv.COLOR_BGR2RGB)

#Akses per piksel
for y in range(image3.shape[0]):
  for x in range(image3.shape[1]):
    blue = image3.item(y,x,0)
    green = image3.item(y,x,1)
    red = image3.item(y,x,2)
    average = (blue + green + red) / 3
    averaging_image3[y,x] = [average, average, average]
    

print('Averaging')
averaging_frame = cv.hconcat((image3, averaging_image3))
plt.imshow(averaging_frame)
```

```python
import matplotlib.pyplot as plt

print('Implementasi Transformasi Grayscale')
print('-----------------------------------')

image3 = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/mandrill.tiff')
lightness_image3 = np.zeros(image3.shape, image3.dtype)

image3 = cv.cvtColor(image3, cv.COLOR_BGR2RGB)

#Akses per piksel
for y in range(image3.shape[0]):
  for x in range(image3.shape[1]):
    blue = image3.item(y,x,0)
    green = image3.item(y,x,1)
    red = image3.item(y,x,2)
    maxBGR = max(blue,green,red)
    minBGR = min(blue, green, red)
    lightness_value = (maxBGR + minBGR) / 2
    lightness_image3[y,x] = [lightness_value, lightness_value, lightness_value]
    
  
print('Lightness')
lightness_frame = cv.hconcat((image3, lightness_image3))
plt.imshow(lightness_frame)
```

**Tampilkanlah warna tertentu pada citra, dan ubah warna lain menjadi grayscale.**

```python
print('Inverse')
print('--------------------------------------------------------------------')

img_gray = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/mandrill.tiff')
img_gray = cv.cvtColor(img_gray,cv.COLOR_BGR2RGB)

temp = np.array(img_gray)
for y in range(img_gray.shape[0]):
    for x in range(img_gray.shape[1]):
        if (temp[y,x,0] < 200 or temp[y,x,1] > 140):
            temp[y,x]= np.mean(temp[y,x])

fig, ax = plt.subplots(1,2)
ax[0].imshow(img_gray)
ax[1].imshow(temp)
```
